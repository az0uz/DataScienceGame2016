# Import Data with user separation

```python
from dsglib import getFile, getFileTrain

data = getFileTrain('data/X_train.csv','data/Y_train.csv')
labelName = 'id,CustomerMD5Key,ReceivedDateTime,SCID,SelectedPackage,FirstDriverMaritalStatus,CarAnnualMileage,CarFuelId,CarUsageId,FirstDriverAge,CarInsuredValue,CarAge,FirstDriverDrivingLicenseNumberY,VoluntaryExcess,CarParkingTypeId,PolicyHolderNoClaimDiscountYears,FirstDriverDrivingLicenceType,CoverIsNoClaimDiscountSelected,CarDrivingEntitlement,CarTransmissionId,SocioDemographicId,PolicyHolderResidencyArea,AllDriversNbConvictions,TodayDate,RatedDriverNumber,IsPolicyholderAHomeowner,CarMakeId,DaysSinceCarPurchase,NameOfPolicyProduct,AffinityCodeId,target'.split(',')
```

# Correlation of same user values

```python
correlated = [0]*31
total = 0
for user in data.values():
    for idx,line in enumerate(user):
        if(idx==0):
            continue
        total += 1
        for idx in range(len(line)):
            if(line[idx]==user[0][idx]):
                correlated[idx] += 1
print correlated
print total
corrScore = map(lambda x: float(x)/total*100,correlated)
for idx,cc in enumerate(corrScore):
    print(labelName[idx]+':\t'+str(cc)+'%')
```

# K-fold split (per user)

```python
foldNumber = 5
foldDict={}
for idx,user in enumerate(data.values()):
    fold = idx%foldNumber
    for row in user:
        foldDict[row[0]] = fold
with open('fold_train.csv', 'w') as f:
    f.write(',fold\n')
    for idx in xrange(2911318):
        fold = foldDict[str(idx)]
        f.write(str(idx)+','+str(fold)+'\n')
```

```python
countDict = {}
for fold in foldDict.values():
    if(not fold in countDict):
        countDict[fold] = 0
    countDict[fold] +=1
print('Number of row per fold: '+str(countDict))
```
