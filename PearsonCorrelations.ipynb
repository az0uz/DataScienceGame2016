```python
liste_Y = ['./Final_results/Y_priv_v5_0_43_.predict',
           './Final_results/Y_priv_v5_1_41_.predict',
           './Final_results/Y_priv_v5_2_39_.predict',
           './Final_results/Y_priv_v5_fold0_40.predict',
           './Final_results/Y_priv_v5_fold2.predict',
           './Final_results/Y_priv_v6FSter_fold2.predict',
           './Final_results/Y_priv_v7fold0.predict',
           './Final_results/Y_priv_v7fold1.predict',
           './Final_results/Y_priv_v7fold2.predict',
           './Final_results/Y_priv_v7fold3.predict',
           './Final_results/Y_priv_v7fold4.predict',
          ]
```

```python
liste_df = []

for f_name in liste_Y :
    liste_df.append(pd.read_csv(f_name, header = None))
```

```python
res = dict()
for i in range(len(liste_Y)) :
    for j in range(len(liste_Y)) :
        if i < j :
            print 'Correlation between ' + liste_Y[i] + ' and ' + liste_Y[j] 
            print repr(float(pearsonr(np.array(liste_df[i].values),np.array(liste_df[j].values))[0]))
            print '----------------------'
            res[liste_Y[i]+'_'+liste_Y[j]] = float(pearsonr(np.array(liste_df[i].values),np.array(liste_df[j].values))[0])
```
